```python
df2["Formatted Date"]= pd.to_datetime(df2["Formatted Date"], utc=True)
yearlist= df2["Formatted Date"].dt.year.unique()
df2
```

```python
plt.figure(figsize=(10,8))
plt.scatter(df2['Temperature (C)'],df2['Humidity'])
plt.xlabel("Temperature" , size=15 , color='g')
plt.ylabel("Humidity", size=15 , color='g')
plt.xticks(rotation=90)
plt.title("UK Weather", size=15, color='r')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.scatter(df2['Wind Speed (km/h)'],df2['Humidity'])
plt.xlabel("Wind Speed (km/h)" , size=15 , color='g')
plt.ylabel("Humidity", size=15 , color='g')
plt.xticks(rotation=90)
plt.title("UK Weather", size=15, color='r')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.scatter(df2['Pressure (millibars)'],df2['Humidity'])
plt.xlabel("Pressure" , size=15 , color='g')
plt.ylabel("Humidity", size=15 , color='g')
plt.xticks(range(0,1000,100),rotation=90)
plt.title("UK Weather", size=15, color='r')
plt.grid()
plt.show()
```

```python
for year in yearlist:
    df8 = df2[df2['Formatted Date'].dt.year == year].copy()
    df8['Month'] = df2.loc[:,'Formatted Date'].dt.month.astype(str)
    figure = plt.figure(figsize=(10 ,8))
    
    hue_order = ['snow','rain']
    countplot = sns.countplot(data=df8,x='Month',hue='Precip Type',hue_order=hue_order, palette='plasma')
    countplot.set_title('Month & Precip Type')
    plt.xlabel("Months of " + str(year))
    plt.savefig(f'D:\\MachinLearning\\ax\\{year}_PrecipType.png')
```
